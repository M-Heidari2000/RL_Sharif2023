## Replay buffer
Typically, people implement replay buffers with one of the following three data structures:

- collections.deque
- list
- numpy.ndarray

**deque** is very easy to handle once you initialize its maximum length (e.g. deque(maxlen=buffer_size)). However, the indexing operation of deque gets terribly slow as it grows up because it is [internally doubly linked list](https://wiki.python.org/moin/TimeComplexity#collections.deque). On the other hands, **list** is an array, so it is relatively faster than deque when you sample batches at every step. Its amortized cost of Get item is [O(1)](https://wiki.python.org/moin/TimeComplexity#list).

Last but not least, let's see **numpy.ndarray**. numpy.ndarray is even faster than list due to the fact that it is [a homogeneous array of fixed-size items](https://docs.scipy.org/doc/numpy/reference/generated/numpy.ndarray.html#numpy.ndarray), so you can get the benefits of [locality of reference](https://en.wikipedia.org/wiki/Locality_of_reference), . Whereas list is an array of pointers to objects, even when all of them are of the same type.

Here, we are going to implement a replay buffer using numpy.ndarray.

Reference: 
- [OpenAI spinning-up](https://github.com/openai/spinningup/blob/master/spinup/algos/sac/sac.py#L10)
- [rainbow-is-all-you-need](https://render.githubusercontent.com/view/ipynb?commit=032d11277cf2436853478a69ca5a4aba03202598&enc_url=68747470733a2f2f7261772e67697468756275736572636f6e74656e742e636f6d2f437572742d5061726b2f7261696e626f772d69732d616c6c2d796f752d6e6565642f303332643131323737636632343336383533343738613639636135613461626130333230323539382f30312e64716e2e6970796e62&nwo=Curt-Park%2Frainbow-is-all-you-need&path=01.dqn.ipynb&repository_id=191133946&repository_type=Repository#Replay-buffer)

```python
class ReplayBuffer:
    """A simple numpy replay buffer."""

    def __init__(self, obs_dim: int, size: int, batch_size: int = 32):
        """Initialize."""
        # ==================================== Your Code (Begin) ====================================
        self.observations = np.zeros((size, obs_dim))
        self.actions = np.zeros((size, 1))
        self.rewards = np.zeros((size, 1))
        self.next_observations = np.zeros((size, obs_dim))
        self.dones = np.zeros((size, 1))
        
        self.batch_size = batch_size
        self.size = size
        
        self._length = 0
        # ==================================== Your Code (End) ====================================


    def store(self,
        obs: np.ndarray,
        act: np.ndarray, 
        rew: float, 
        next_obs: np.ndarray,
        done: bool,
    ):
        """Store the transition in buffer."""
        # ==================================== Your Code (Begin) ====================================
        self.observations[1:, :] = self.observations[:-1, :]
        self.observations[0, :] = obs
        
        self.actions[1:] = self.actions[:-1]
        self.actions[0] = act
        
        self.rewards[1:] = self.rewards[:-1]
        self.rewards[0] = rew
        
        self.next_observations[1:, :] = self.next_observations[:-1, :]
        self.next_observations[0, :] = next_obs
        
        self.dones[1:] = self.dones[:-1]
        self.dones[0] = 1 if done is True else 0
        
        self._length += 1
        self._length = min(self._length, self.size)
        # ==================================== Your Code (End) ====================================


    def sample_batch(self) -> Dict[str, np.ndarray]:
        """Randomly sample a batch of experiences from memory."""
        # ==================================== Your Code (Begin) ====================================
        sample_indices = np.random.choice(range(self.size), size=self.batch_size, replace=False)
        
        batch_data = {
            'observations': self.observations[sample_indices, :],
            'actions': self.actions[sample_indices],
            'rewards': self.rewards[sample_indices],
            'next_observations': self.next_observations[sample_indices, :],
            'dones': self.dones[sample_indices]
        }
        
        return batch_data
        # ==================================== Your Code (End) ====================================


    def __len__(self) -> int:
        # ==================================== Your Code (Begin) ====================================
        return self._length
        # ==================================== Your Code (End) ====================================
```

```python
# Test for the `ReplayBuffer`

replay_buffer = ReplayBuffer(
    obs_dim=1,
    size=2,
    batch_size=2
)

replay_buffer.store(
    obs=1,
    act=1,
    rew=1,
    next_obs=2,
    done=False
)

sample1 = replay_buffer.sample_batch()
print(sample1, end='\n\n')

replay_buffer.store(
    obs=2,
    act=2,
    rew=2,
    next_obs=3,
    done=False
)

sample2 = replay_buffer.sample_batch()
print(sample2, end='\n\n')

replay_buffer.store(
    obs=3,
    act=3,
    rew=3,
    next_obs=4,
    done=True
)

sample3 = replay_buffer.sample_batch()
print(sample3)
```

## OU Noise
**Ornstein-Uhlenbeck** process generates temporally correlated exploration, and it effectively copes with physical control problems of inertia.

$$
dx_t = \theta(\mu - x_t) dt + \sigma dW_t
$$

Reference: 
- [Udacity github](https://github.com/udacity/deep-reinforcement-learning/blob/master/ddpg-pendulum/ddpg_agent.py)
- [Wiki](https://en.wikipedia.org/wiki/Ornstein%E2%80%93Uhlenbeck_process)

```python
class OUNoise:
    """Ornstein-Uhlenbeck process.
    Taken from Udacity deep-reinforcement-learning github repository:
    https://github.com/udacity/deep-reinforcement-learning/blob/master/
    ddpg-pendulum/ddpg_agent.py
    """

    def __init__(
        self, 
        size: int, 
        mu: float = 0.0, 
        theta: float = 0.15, 
        sigma: float = 0.2,
    ):
        """Initialize parameters and noise process."""
        self.state = np.float64(0.0)
        self.mu = mu * np.ones(size)
        self.theta = theta
        self.sigma = sigma
        self.reset()

    def reset(self):
        """Reset the internal state (= noise) to mean (mu)."""
        self.state = copy.copy(self.mu)

    def sample(self) -> np.ndarray:
        """Update internal state and return it as a noise sample."""
        x = self.state
        dx = self.theta * (self.mu - x) + self.sigma * np.array(
            [random.random() for _ in range(len(x))]
        )
        self.state = x + dx
        return self.state
```

# PPO

- PPO: [J. Schulman et al., "Proximal Policy Optimization Algorithms." arXiv preprint arXiv:1707.06347, 2017.](https://arxiv.org/abs/1707.06347.pdf)
- TRPO: [Schulman, John, et al. "Trust region policy optimization." International conference on machine learning. 2015.](http://proceedings.mlr.press/v37/schulman15.pdf)

There are two kinds of algorithms of PPO: PPO-Penalty and PPO-Clip. Here, we'll implement PPO-clip version.

TRPO computes the gradients with a complex second-order method. On the other hand, PPO tries to solve the problem with a first-order methods that keep new policies close to old. To simplify the surrogate objective, let $r(\theta)$ denote the probability ratio

$$ L^{CPI}(\theta) = \hat {\mathbb{E}}_t \left [ {\pi_\theta(a_t|s_t) \over \pi_{\theta_{old}}(a_t|s_t)} \hat A_t\right] = \hat {\mathbb{E}}_t \left [ r_t(\theta) \hat A_t \right ].$$

The objective is penalized further away from $r_t(\theta)$

$$ L^{CLIP}(\theta)=\hat {\mathbb{E}}_t \left [ \min(r_t(\theta) \hat A_t, \text{clip}(r_t(\theta), 1-\epsilon, 1+\epsilon)\hat A_t) \right ] $$

If the advantage is positive, the objective will increase. As a result, the action becomes more likely. If advantage is negative, the objective will decrease. AS a result, the action becomes less likely.

# DDPG 

[T. P. Lillicrap et al., "Continuous control with deep reinforcement learning." arXiv preprint arXiv:1509.02971, 2015.](https://arxiv.org/pdf/1509.02971.pdf)

Deep Q Network(DQN)([Mnih et al., 2013;2015](https://storage.googleapis.com/deepmind-media/dqn/DQNNaturePaper.pdf)) algorithm is combined advances in deep learning with reinforcement learning. However, while DQN solves problems with high-dimentional observation spaces, it can only handle discrete and low-dimentional action spaces because of using greedy policy. For learning in high-dimentional and continous action spaces, the authors combine the actor-critic approach with insights from the recent success of DQN. Deep DPG(DDPG) is based on the deterministic policy gradient(DPG) algorithm ([Silver et al., 2014](http://proceedings.mlr.press/v32/silver14.pdf)). 

### Deterministic policy gradient
The DPG algorithm maintains a parameterized actor function $\mu(s|\theta^{\mu})$ which specifies the current policy by deterministically mapping states to a specific action. The critic $Q(s, a)$ is learned using the Bellman equation as in Q-learning. The actor is updated by following the applying the chain rule to the expected return from the start distribution $J$ with respect to the actor parameters

$$
\begin{align*}
\nabla_{\theta^{\mu}}J &\approx E_{s_t\sim\rho^\beta} [\nabla_{\theta^{\mu}} Q(s,a|\theta^Q)|_{s=s_t, a=\mu(s_t|\theta^\mu)}] \\
&= E_{s_t\sim\rho^\beta} [\nabla_{a} Q(s,a|\theta^Q)|_{s=s_t, a=\mu(s_t)} \nabla_{\theta^{\mu}} \mu(s|\theta^\mu)|_{s=s_t}]
\end{align*}
$$

### Soft update target network
Since the network $(Q(s,a|\theta^Q)$ being updated is also used in calculating the target value, the Q update is prone to divergence. To avoid this, the authors use **the target network** like DQN, but modified for actor-critic and using **soft target updates**. Target netwokrs is created by copying the actor and critic networks, $Q'(s,a|\theta^{Q'})$ and $\mu'(s|\theta^{\mu`})$ respectively, that are used for calculating the target values. The weights of these target networks are then updated by having them slowly track the learned networks:

$$
\theta' \leftarrow \tau \theta + (1 - \tau)\theta' \ \ \ {with} \ \tau \ll 1.
$$

It greatly improves the stability of learning.

### Exploration for continuous action space
An advantage of off-policy algorithms such as DDPG is that we can treat the problem of exploration independently from the learning algorithm. The authors construct an exploration policy $\mu'$ by adding noise sampled from a noise process $\mathcal{N}$ to the actor policy

$$
\mu'(s_t) = \mu(s_t|\theta^{\mu}_t) + \mathcal{N}
$$

$\mathcal{N}$ can be chosen to suit the environment. The authors used **Ornstein-Uhlenbeck process** to generate temporally correlated exploration.

# Networks


## PPO Networks

We will use two separated networks for actor and critic respectively. The actor network consists of two fully connected hidden layer with ReLU branched out two fully connected output layers for mean and standard deviation of Gaussian distribution. Pendulum-v0 has only one action which has a range from -2 to 2. In order to fit the range, the actor outputs the mean value with tanh. The result will be scaled in ActionNormalizer class. On the one hand, the critic network has three fully connected layers as two hidden layers (ReLU) and an output layer. One thing to note is that we initialize the last layers' weights and biases as uniformly distributed.

```python
def init_layer_uniform(layer: nn.Linear, init_w: float = 3e-3) -> nn.Linear:
    """Init uniform parameters on the single layer."""
    # ==================================== Your Code (Begin) ====================================
    nn.init.uniform_(layer.weight, -init_w, init_w)
    nn.init.uniform_(layer.bias, -init_w, init_w)
    # ==================================== Your Code (End) ====================================


class PPOActor(nn.Module):
    def __init__(
        self,
        in_dim: int, 
        out_dim: int, 
        log_std_min: int = -20,
        log_std_max: int = 0,
        init_w: float = 3e-3
    ):
        """Initialize."""
        self.log_std_min = log_std_min
        self.log_std_max = log_std_max
        
        super().__init__()

        # ==================================== Your Code (Begin) ====================================
        # Common body of the network
        self.body = nn.Sequential(
            nn.Linear(in_dim, 64),
            nn.ReLU(),
            nn.Linear(64, 64),
            nn.ReLU(),
        )
        
        # mean head
        self.mean_head = nn.Linear(64, 1)
        init_layer_uniform(self.mean_head, init_w=init_w)
        
        # standard deviation head
        self.std_head = nn.Linear(64, 1)
        init_layer_uniform(self.std_head, init_w=init_w)
        
        # ==================================== Your Code (End) ====================================


    def forward(self, state: torch.Tensor) -> torch.Tensor:
        """Forward method implementation."""
        # ==================================== Your Code (Begin) ====================================
        # return selected action and a normal distribution as output of this module
        body_result = self.body(state)
        mean = torch.tanh(self.mean_head(body_result))
        log_std_raw = torch.tanh(self.std_head(body_result))
        log_stdـscaled = log_std_raw * 0.5 * (self.log_std_max - self.log_std_min) \
                        + 0.5 * (self.log_std_max + self.log_std_min)
        std_scaled = torch.exp(log_stdـscaled)
        dist = Normal(loc=mean, scale=std_scaled)
        action = dist.sample()
        
        return action, dist
        # ==================================== Your Code (End) ====================================


class PPOCritic(nn.Module):
    def __init__(self, in_dim: int, init_w: float=3e-3):
        """Initialize."""
        super().__init__()
        # ==================================== Your Code (Begin) ====================================
        self.body = nn.Sequential(
            nn.Linear(in_dim, 64),
            nn.ReLU(),
            nn.Linear(64, 64),
            nn.ReLU(),
        )
        
        self.head = nn.Linear(64, 1)
        init_layer_uniform(self.head, init_w=init_w)
        # ==================================== Your Code (End) ====================================


    def forward(self, state: torch.Tensor) -> torch.Tensor:
        """Forward method implementation."""
        # ==================================== Your Code (Begin) ====================================
        body_results = self.body(state)
        estimated_value = self.head(body_results)
        return estimated_value
        # ==================================== Your Code (End) ====================================
```

## DDPG Networks
We are going to use two separated networks for actor and critic. The actor network has three fully connected layers and three non-linearity functions, **ReLU** for hidden layers and **tanh** for the output layer. On the other hand, the critic network has three fully connected layers, but it used two activation functions for hidden layers **ReLU**. Plus, its input sizes of critic network are sum of state sizes and action sizes. One thing to note is that we initialize the final layer's weights and biases so that they are **uniformly distributed.**

```python
class DDPGActor(nn.Module):
    def __init__(
        self, 
        in_dim: int, 
        out_dim: int,
        init_w: float = 3e-3,
    ):
        """Initialize."""
        super().__init__()
        
        # ==================================== Your Code (Begin) ====================================
        # 1. set the hidden layers
        # 2. init hidden layers uniformly
        
        self.body = nn.Sequential(
            nn.Linear(in_dim, 32),
            nn.ReLU(),
            nn.Linear(32, 32),
            nn.ReLU()
        )
        
        self.head = nn.Linear(32, out_dim)
        init_layer_uniform(self.head, init_w=init_w)
        
        # ==================================== Your Code (End) ====================================

    def forward(self, state: torch.Tensor) -> torch.Tensor:
        """Forward method implementation."""
        # ==================================== Your Code (Begin) ====================================
        # use a tanh function as a ativation function for output layer
        body_result = self.body(state)
        action = torch.tanh(self.head(body_result))
        return action
        # ==================================== Your Code (End) ====================================
    
    
class DDPGCritic(nn.Module):
    def __init__(
        self, 
        in_dim: int, 
        init_w: float = 3e-3,
    ):
        """Initialize."""
        super().__init__()
        
        # ==================================== Your Code (Begin) ====================================
        # 1. set the hidden layers
        # 2. init hidden layers uniformly
        
        self.body = nn.Sequential(
            nn.Linear(in_dim, 32),
            nn.ReLU(),
            nn.Linear(32, 32),
            nn.ReLU()
        )
        
        self.head = nn.Linear(32, 1)
        init_layer_uniform(self.head, init_w=init_w)
        # ==================================== Your Code (End) ====================================

    def forward(
        self, state: torch.Tensor, action: torch.Tensor
    ) -> torch.Tensor:
        """Forward method implementation."""
        # ==================================== Your Code (Begin) ====================================
        # notice that this value function is Q(s, a)
        network_input = torch.cat((state, action), dim=1)
        body_result = self.body(network_input)
        estimated_value = self.head(body_result)
        
        return estimated_value
        # ==================================== Your Code (End) ====================================
```

# Agents 

## PPO Agent
Here is a summary of PPOAgent class.

| Method           | Note                                                 |
|---               |---                                                   |
|select_action     | select an action from the input state.               |
|step              | take an action and return the response of the env.   |
|update_model      | update the model by gradient descent.                |
|train             | train the agent during num_frames.                   |
|test              | test the agent (1 episode).                          |
|_plot             | plot the training progresses.                        |


PPO updates the model several times(`epoch`) using the stacked memory. By `ppo_iter` function, It yield the samples of stacked memory by interacting a environment.

```python
def ppo_iter(
    epoch: int,
    mini_batch_size: int,
    states: torch.Tensor,
    actions: torch.Tensor,
    values: torch.Tensor,
    log_probs: torch.Tensor,
    returns: torch.Tensor,
    advantages: torch.Tensor,
):
    """Yield mini-batches."""
    
    # ==================================== Your Code (Begin) ====================================
    rollout_length = advantages.shape[0]
    
    for _ in range(epoch):
        for _ in range(rollout_length // mini_batch_size):
            indices = np.random.choice(rollout_length, mini_batch_size)
            yield (
                states[indices, :],
                actions[indices],
                values[indices],
                log_probs[indices],
                returns[indices],
                advantages[indices]
            )
    # ==================================== Your Code (End) ====================================
```

```python
class PPOAgent:
    """PPO Agent.
    Attributes:
        env (gym.Env): Gym env for training
        gamma (float): discount factor
        tau (float): lambda of generalized advantage estimation (GAE)
        batch_size (int): batch size for sampling
        epsilon (float): amount of clipping surrogate objective
        epoch (int): the number of update
        rollout_len (int): the number of rollout
        entropy_weight (float): rate of weighting entropy into the loss function
        actor (nn.Module): target actor model to select actions
        critic (nn.Module): critic model to predict state values
        transition (list): temporory storage for the recent transition
        device (torch.device): cpu / gpu
        total_step (int): total step numbers
        is_test (bool): flag to show the current mode (train / test)        
    """

    def __init__(
        self,
        env: gym.Env,
        batch_size: int,
        gamma: float,
        tau: float,
        epsilon: float,
        epoch: int,
        rollout_len: int,
        entropy_weight: float,
    ):
        """Initialize."""
        # ==================================== Your Code (Begin) ====================================
        # 1. set hyperparameters
        self.env = env
        self.batch_size = batch_size
        self.gamma = gamma
        self.tau = tau
        self.epsilon = epsilon
        self.epoch = epoch
        self.rollout_len = rollout_len
        self.entropy_weight = entropy_weight
        
        # observation space and action space dimensions
        self.observation_dim = env.observation_space.shape[0]
        self.action_dim = env.action_space.shape[0]
        
        # 2. check device: cpu/GPU
        self.device = 'cuda' if torch.cuda.is_available() else 'cpu'
        
        # 3. init actor critic networks
        # Actor
        self.actor = PPOActor(
            in_dim=self.observation_dim,
            out_dim=self.action_dim
        ).to(self.device)
        # Critic
        self.critic = PPOCritic(
            in_dim=self.observation_dim
        ).to(self.device)
        
        # 4. set Optimizer for each network
        actor_lr = 1e-6
        critic_lr = 1e-6
        self.actor_optimizer = torch.optim.Adam(self.actor.parameters(), lr=actor_lr)
        self.critic_optimizer = torch.optim.Adam(self.critic.parameters(), lr=critic_lr)
        
        # 5. consider memory for training
        self.rollout_observations: List[torch.tensor] = []
        self.rollout_actions: List[torch.tensor] = []
        self.rollout_log_probs: List[torch.tensor] = []
        self.rollout_rewards: List[torch.tensor] = []
        self.rollout_dones: List[torch.tensor] = []
        
        # 6. set total step counts equal to 1
        self.total_step = 1
        
        # 7. define a mode for train/test
        self.is_test = False
        # ==================================== Your Code (End) ====================================
        
    def select_action(self, state: np.ndarray) -> np.ndarray:
        """Select an action from the input state."""
        # ==================================== Your Code (Begin) ====================================
        # 1. select action for train or test mode
        state = torch.tensor(state).to(self.device)
        
        with torch.no_grad():
            action, dist = self.actor(state)
            selected_action = dist.mean if self.is_test else action
            action_log_prob = dist.log_prob(action)
            
            # 2. if you are in train mode take care of filing considered memory
            if not self.is_test:
                self.rollout_observations.append(state)
                self.rollout_actions.append(selected_action.item())
                self.rollout_log_probs.append(action_log_prob.item())
        
            return selected_action.cpu().item()
        # ==================================== Your Code (End) ====================================

    def step(self, action: np.ndarray) -> Tuple[np.ndarray, np.float64, bool]:
        """Take an action and return the response of the env."""
        # ==================================== Your Code (Begin) ====================================
        next_state, reward, _, truncated, _ = self.env.step(action)
        
        if not self.is_test:
            self.rollout_rewards.append(reward)
            self.rollout_dones.append(truncated)
        
        return next_state, reward, truncated
        # ==================================== Your Code (End) ====================================

    def update_model(
        self, next_state: np.ndarray
    ) -> Tuple[torch.Tensor, torch.Tensor]:
        """Update the model by gradient descent."""
        # ==================================== Your Code (Begin) ====================================
        # 1. set device
        self.device = 'cuda' if torch.cuda.is_available() else 'cpu'
        next_state = next_state.reshape(1, -1)
        next_state = torch.tensor(next_state).to(self.device)
        next_value = self.critic(next_state)
        
        network_inputs = torch.concat(self.rollout_observations, dim=0)
        self.rollout_values = self.critic(network_inputs).detach().flatten()
        
        self.rollout_returns, self.rollout_advantages = self.compute_gae(
            next_value=next_value.item(),
            rewards=self.rollout_rewards,
            dones=self.rollout_dones,
            values=[x.item() for x in self.rollout_values],
            gamma=self.gamma,
            tau=self.tau
        )
                
        with torch.no_grad():
            self.rollout_observations = torch.concat(self.rollout_observations, dim=0).to(self.device)
            self.rollout_actions = torch.tensor(self.rollout_actions).to(self.device)
            self.rollout_log_probs = torch.tensor(self.rollout_log_probs).to(self.device)
            self.rollout_returns = torch.tensor(self.rollout_returns).to(self.device)
            self.rollout_advantages = torch.tensor(self.rollout_advantages).to(self.device)
            
        actor_losses = []
        critic_losses = []
                
        # 2. for each step:
        for (
            batch_states,
            batch_actions,
            batch_values,
            batch_log_probs,
            batch_returns,
            batch_advantages
        ) in ppo_iter(
            epoch=self.epoch,
            mini_batch_size=self.batch_size,
            states=self.rollout_observations,
            actions=self.rollout_actions,
            values=self.rollout_values,
            log_probs=self.rollout_log_probs,
            returns=self.rollout_returns,
            advantages=self.rollout_advantages
        ):
                                                            
            _, dists = self.actor(batch_states)
            log_probs = dists.log_prob(batch_actions.reshape(-1, 1)).flatten()
                        
            # 3.    calculate ratios
            ratios = torch.exp(log_probs - batch_log_probs)
                    
            # 4.    calculate actor_loss
            surr_loss = ratios * batch_advantages
            clipped_surr_loss = torch.clamp(ratios, 1-self.epsilon, 1+self.epsilon) * batch_advantages
                        
            # 5.    calculate entropy
            entropy = dists.entropy().mean()
            actor_loss = - (torch.min(surr_loss, clipped_surr_loss).mean() + self.entropy_weight * entropy)

            # 6.    calculate critic_loss
            values = self.critic(batch_states)
            critic_loss = ((batch_returns - values) ** 2).mean()
            
            # 7.    Train  critic
            self.critic_optimizer.zero_grad()
            critic_loss.backward()
            self.critic_optimizer.step()

            # 8.    Train actor
            self.actor_optimizer.zero_grad()
            actor_loss.backward()
            self.actor_optimizer.step()

            actor_losses.append(actor_loss.item())
            critic_losses.append(critic_loss.item())
            
        self.rollout_observations = []
        self.rollout_actions = []
        self.rollout_rewards = []
        self.rollout_values = []
        self.rollout_masks = []
        self.rollout_log_probs = []
                
        actor_loss = np.mean(actor_losses)
        critic_loss = np.mean(critic_losses)
        
        return actor_loss, critic_loss
        
        # ==================================== Your Code (End) ====================================
    
    def train(self, num_frames: int, plotting_interval: int = 200):
        """Train the agent."""
        # ==================================== Your Code (Begin) ====================================
        # 1. set the status of trainig
        self.is_test = False
        
        # 2. Reset environment
        state, info = self.env.reset()
        
        actor_losses = []
        critic_losses = []
        scores = []
        score = 0
        
        # 3. for number of frames:
        while self.total_step <= num_frames + 1:
            for _ in range(self.rollout_len):
                self.total_step += 1
                
                # 4. select an action
                state = state.reshape(1, -1)
                action = self.select_action(state)
                # 5. step in environment
                next_state, reward, truncated = self.step(action)
                
                state = next_state
                score += reward
                
                # Start a new episode
                if truncated:
                    state, info = self.env.reset()
                    scores.append(score)
                    score = 0
                    
            
            # 6. update model
            actor_loss, critic_loss = self.update_model(next_state)
            print(f'actor loss: {actor_loss}')
            print(f'critic loss: {critic_loss}')
            print(f'score: {scores[-1]}')
            print()
            actor_losses.append(actor_loss)
            critic_losses.append(critic_loss)
        
        # 7. terminate environment after training is finished
        
        self._plot(
            self.total_step,
            scores,
            actor_losses,
            critic_losses
        )
        
        self.env.close()
        # ==================================== Your Code (End) ====================================

    def test(self):
        """Test the agent."""
        # ==================================== Your Code (Begin) ====================================
        # 1. set the status of trainig
        self.is_test = True
        
        with torch.no_grad():
            # 2. Reset environment
            state, info = env.reset(seed=seed)
            # 3. roll out one episode living in the environment and save frames for getting render
            score = 0
            truncated = False
            frames = []

            while not truncated:
                frames.append(env.render())
                action = self.select_action(state=state)
                next_state, reward, truncated = self.step(action=action)
                state = next_state
                score += reward
                
            print(score)

            return frames
        # ==================================== Your Code (End) ====================================

        
    def compute_gae(
        self,
        next_value: float,
        rewards: list,
        dones: list,
        values: list,
        gamma: float,
        tau: float
    ) -> (List, List):
        """Compute gae"""
                
        with torch.no_grad():
        
            values.append(next_value)
            gae = 0
            R = 0
            advantages = []
            returns = []

            for t in range(len(rewards)-1, -1, -1):
                # Compute reward to go
                R = rewards[t] + (1 - dones[t]) * gamma * R
                returns.append(R.item())

                # Compute Advantage
                delta = rewards[t] + gamma * (1 - dones[t]) * values[t+1] - values[t]
                gae = delta + gamma * tau * (1 - dones[t]) * gae
                advantages.append(gae.item())

            advantages = list(reversed(advantages))
            returns = list(reversed(returns))

            return returns, advantages

    def _plot(
        self,
        frame_idx: int,
        scores: List[float],
        actor_losses: List[float],
        critic_losses: List[float],
    ):
        """Plot the training progresses."""
        # ==================================== Your Code (Begin) ====================================
        # 1. define a function for sub plots
        # 2. for each variable plot the specific subplot
        # 3. plot the whole figure
        
        
        fig = plt.figure(figsize=(12, 3))
        ax1 = fig.add_subplot(1, 3, 1)
        ax2 = fig.add_subplot(1, 3, 2)
        ax3 = fig.add_subplot(1, 3, 3)
        
        clear_output(True)
        
        ax1.plot(scores)
        ax1.set_title('score')
        
        ax2.plot(critic_losses)
        ax2.set_title('critic loss')
        
        ax3.plot(actor_losses)
        ax3.set_title('actor loss')
        
        plt.show()
        # ==================================== Your Code (End) ====================================
```

## DDPG Agent
Here is a summary of DDPGAgent class.

| Method           | Note                                                 |
|---               |---                                                  |
|select_action     | select an action from the input state.               |
|step              | take an action and return the response of the env.   |
|update_model      | update the model by gradient descent.                |
|train             | train the agent during num_frames.                   |
|test              | test the agent (1 episode).                          |
|\_target_soft_update| soft update from the local model to the target model.|
|\_plot              | plot the training progresses.                        |

```python
class DDPGAgent:
    """DDPGAgent interacting with environment.
    
    Attribute:
        + env (gym.Env): openAI Gym environment
        + actor (nn.Module): target actor model to select actions
        + actor_target (nn.Module): actor model to predict next actions
        actor_optimizer (Optimizer): optimizer for training actor
        + critic (nn.Module): critic model to predict state values
        + critic_target (nn.Module): target critic model to predict state values
        critic_optimizer (Optimizer): optimizer for training critic
        + memory (ReplayBuffer): replay memory to store transitions
        + batch_size (int): batch size for sampling
        + gamma (float): discount factor
        + tau (float): parameter for soft target update
        + initial_random_steps (int): initial random action steps
        + noise (OUNoise): noise generator for exploration
        + device (torch.device): cpu / gpu
        + transition (list): temporory storage for the recent transition
        + total_step (int): total step numbers
        + is_test (bool): flag to show the current mode (train / test)
    """
    def __init__(
        self,
        env: gym.Env,
        memory_size: int,
        batch_size: int,
        ou_noise_theta: float,
        ou_noise_sigma: float,
        gamma: float = 0.99,
        tau: float = 5e-3,
        initial_random_steps: int = 1e4,
    ):
        """Initialize."""

        # ==================================== Your Code (Begin) ====================================
        # 1. initialize hyper parameters, reply buffer and environment
        self.env = env
        self.gamma = gamma
        self.tau = tau
        self.initial_random_steps = initial_random_steps
        self.batch_size = batch_size
        
        self.observation_dim = env.observation_space.shape[0]
        self.action_dim = env.action_space.shape[0]
        
        self.memory = ReplayBuffer(
            obs_dim=self.observation_dim,
            size=memory_size,
            batch_size=batch_size
        )
        
        # 2. set device
        self.device = 'cuda' if torch.cuda.is_available() else 'cpu'
        
        # 3. set target entropy, log alpha and alpha optimizer ?????
        
        # 4. init actor network
        self.actor = DDPGActor(
            in_dim=self.observation_dim,
            out_dim=self.action_dim
        ).to(self.device)
        
        self.actor_target = DDPGActor(
            in_dim=self.observation_dim,
            out_dim=self.action_dim
        ).to(self.device)
        
        self.actor_target.load_state_dict(self.actor.state_dict())
        
        # 5. init value fuction (value critic)
        self.critic = DDPGCritic(
            in_dim=self.observation_dim + self.action_dim
        ).to(self.device)
        
        self.critic_target = DDPGCritic(
            in_dim=self.observation_dim + self.action_dim
        ).to(self.device)
        
        self.critic_target.load_state_dict(self.critic.state_dict())
        
        # 6. init OUNoise
        self.noise = OUNoise(
            size=self.action_dim,
            theta=ou_noise_theta,
            sigma=ou_noise_sigma
        )
        
        # 7. set Optimizers
        actor_lr = 1e-4
        critic_lr = 1e-3
        self.actor_optimizer = torch.optim.Adam(self.actor.parameters(), lr=actor_lr)
        self.critic_optimizer = torch.optim.Adam(self.critic.parameters(), lr=critic_lr)
        
        # consider stroring transitions in memeory, counting steps and specify train/test mode
        self.transition = []
        self.total_steps = 1
        self.is_test = False
        # ==================================== Your Code (End) ====================================
    
    def select_action(self, state: np.ndarray) -> np.ndarray:
        """Select an action from the input state."""
        
        # ==================================== Your Code (Begin) ====================================
        state = torch.tensor(state).to(self.device)
        
        # 1. check if initial random action should be conducted
        with torch.no_grad():
            
            if self.is_test:
                action = self.actor_target(state).cpu().item()
                return action
            
            if self.total_steps < self.initial_random_steps:
                action = self.env.action_space.sample()
            else:
                action = self.actor(state).cpu().item()
            # 2. add noise for exploration during training
            if not self.is_test:
                action = action + self.noise.sample()
                action = np.clip(action, a_min=-1, a_max=1)
                # 3. store transition
                self.transition.append(state.cpu().numpy())
                self.transition.append(action)
                
            # return selected action
            return action
        # ==================================== Your Code (End) ====================================
    
    def step(self, action: np.ndarray) -> Tuple[np.ndarray, np.float64, bool]:
        """Take an action and return the response of the env."""
        # ==================================== Your Code (Begin) ====================================
        # step in environment and save transition in memory if you are not in test mode
        next_state, reward, _, truncated, _ = self.env.step(action)
        
        if not self.is_test:
            self.transition.append(reward)
            self.transition.append(next_state)
            self.transition.append(truncated)
            # store the transition in the buffer
            self.memory.store(*self.transition)
            self.transition = []
            
        return next_state, reward, truncated
        # ==================================== Your Code (End) ====================================
    
    def update_model(self) -> torch.Tensor:
        """Update the model by gradient descent."""
        # ==================================== Your Code (Begin) ====================================
        # 1. set device
        self.device = 'cuda' if torch.cuda.is_available() else 'cpu'
        
        # 2. get a batch from memory and calculate the return
        batch = self.memory.sample_batch()
        states = torch.FloatTensor(batch['observations']).to(self.device)
        actions = torch.FloatTensor(batch['actions']).to(self.device)
        rewards = torch.FloatTensor(batch['rewards']).to(self.device)
        next_states = torch.FloatTensor(batch['next_observations']).to(self.device)
        dones = torch.FloatTensor(batch['dones']).to(self.device)
                
        # 3. calculate the loss for actor and critic networks
        next_actions = self.actor_target(next_states)
        targets = rewards + self.gamma * (1 - dones) * self.critic_target(next_states, next_actions).detach()
                        
        # 4. update target
        critic_loss = ((targets - self.critic(states, actions)) ** 2).mean()
        
        self.critic_optimizer.zero_grad()
        critic_loss.backward()
        self.critic_optimizer.step()
        
        actor_loss = -self.critic(states, self.actor(states)).mean()
        
        self.actor_optimizer.zero_grad()
        actor_loss.backward()
        self.actor_optimizer.step()
        
        # 5. Soft update target networks
        self._target_soft_update()
        
        return actor_loss.item(), critic_loss.item()
        # ==================================== Your Code (End) ====================================
    
    def train(self, num_frames: int, plotting_interval: int = 200):
        """Train the agent."""
        # ==================================== Your Code (Begin) ====================================
        # 1. set the status of trainig
        self.is_test = False
        
        # 2. Reset environment
        state, info = self.env.reset()
        
        critic_losses = []
        actor_losses = []
        scores = []
        score = 0
        
        # 3. for number of frames:
        while self.total_steps <= num_frames + 1:
            self.total_steps += 1
        
            # 4.    select an action
            state = state.reshape(1, -1)
            action = self.select_action(state)
            # 5.    step in environment
            next_state, reward, truncated = self.step(action)
            state = next_state
            score += reward
            
            # Start a new episode
            if truncated:
                state, info = self.env.reset()
                scores.append(score)
                if len(critic_losses) > 0:
                    print(f'critic loss: {np.mean(critic_losses)}')
                    print(f'actor loss: {np.mean(actor_losses)}')
                    print(f'score: {score}')
                score = 0
                    
            # 6.    update model
            if self.total_steps > self.initial_random_steps and \
            len(self.memory) >= self.batch_size:
                actor_loss, critic_loss = self.update_model()
                actor_losses.append(actor_loss)
                critic_losses.append(critic_loss)
            
        # 8. terminate environment after training is finished    
        # 7. plot the computed variables
        self._plot(
            self.total_steps,
            scores,
            actor_losses,
            critic_losses
        )
        
        self.env.close()
        # ==================================== Your Code (End) ====================================
        
    def test(self):
        """Test the agent."""
        # ==================================== Your Code (Begin) ====================================
        # 1. set the status of trainig
        self.is_test = True
        
        with torch.no_grad():
            # 2. Reset environment
            state, info = env.reset(seed=seed)
            # 3. roll out one episode living in the environment and save frames for getting render
            score = 0
            truncated = False
            frames = []

            while not truncated:
                frames.append(env.render())
                action = self.select_action(state=state)
                next_state, reward, truncated = self.step(action=action)
                state = next_state
                score += reward
                
            print(score)

            return frames
        # ==================================== Your Code (End) ====================================
    
    def _target_soft_update(self):
        """Soft-update: target = tau*local + (1-tau)*target."""
        # ==================================== Your Code (Begin) ====================================
        # Soft update actor
        target_actor_state_dict = self.actor_target.state_dict()
        actor_state_dict = self.actor.state_dict()
        
        for key in actor_state_dict:
            target_actor_state_dict[key] = self.tau * actor_state_dict[key] + \
                (1 - self.tau) * target_actor_state_dict[key]
        
        self.actor_target.load_state_dict(target_actor_state_dict)
        
        # Soft update critic
        target_critic_state_dict = self.critic_target.state_dict()
        critic_state_dict = self.critic.state_dict()
        
        for key in critic_state_dict:
            target_critic_state_dict[key] = self.tau * critic_state_dict[key] + \
                (1 - self.tau) * target_critic_state_dict[key]
        
        self.critic_target.load_state_dict(target_critic_state_dict)
        
        # ==================================== Your Code (End) ====================================
    
    def _plot(
        self, 
        frame_idx: int, 
        scores: List[float], 
        actor_losses: List[float], 
        critic_losses: List[float], 
    ):
        """Plot the training progresses."""
        # ==================================== Your Code (Begin) ====================================
        fig = plt.figure(figsize=(12, 3))
        ax1 = fig.add_subplot(1, 3, 1)
        ax2 = fig.add_subplot(1, 3, 2)
        ax3 = fig.add_subplot(1, 3, 3)
        
        clear_output(True)
        
        ax1.plot(scores)
        ax1.set_title('score')
        
        ax2.plot(critic_losses)
        ax2.set_title('critic loss')
        
        ax3.plot(actor_losses)
        ax3.set_title('actor loss')
        
        plt.show()
        # ==================================== Your Code (End) ====================================
```

# Environment
*ActionNormalizer* is an action wrapper class to normalize the action values ranged in (-1. 1). Thanks to this class, we can make the agent simply select action values within the zero centered range (-1, 1).

```python
class ActionNormalizer(gym.ActionWrapper):
    """Rescale and relocate the actions."""

    def __init__(self, env, low, high):
        super().__init__(env)
        self.low = low
        self.high = high
    
    def action(self, action: np.ndarray) -> np.ndarray:
        """Change the range (-1, 1) to (low, high)."""
        # ==================================== Your Code (Begin) ====================================
        scaled_action = ((self.high - self.low) * action + (self.high + self.low)) / 2
        return scaled_action
        # ==================================== Your Code (End) ====================================


    def reverse_action(self, action: np.ndarray) -> np.ndarray:
        """Change the range (low, high) to (-1, 1)."""
        # ==================================== Your Code (Begin) ====================================
        scaled_action = (2 * action - (self.high + self.low)) / (self.high - self.low)
        return scaled_action
        # ==================================== Your Code (End) ====================================
```
